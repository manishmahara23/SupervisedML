Merging True and Fake Dataframes

```python
df_merge = pd.concat([df_fake, df_true], axis =0 )
df_merge.head()
```

```python
df_merge.columns
```

Removing columns which are not required

```python
df = df_merge.drop(["title", "subject","date"], axis = 1)
```

```python
df.isnull().sum()
```

Random Shuffling the dataframe

```python
df = df.sample(frac = 1)
```

```python
df.head()
```

```python
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

Creating a function to process the texts

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

Defining dependent and independent variables

```python
x = df["text"]
y = df["class"]
```

Splitting Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

Convert text to vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

Model Testing

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)

    print("\n\nLR Prediction: {} \nDT Prediction: {}".format(output_lable(pred_LR[0]),output_lable(pred_DT[0])))

# # Example usage of manual_testing with a fake news article
# news_text_fake = "Donald Trump says he will build a wall and make Mexico pay for it."
# manual_testing(news_text_fake)

# # Example usage of manual_testing with a true news article
# news_text_true = "The Supreme Court is the highest judicial body in the United States."
# manual_testing(news_text_true)
```
