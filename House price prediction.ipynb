Splliting the dataset

```python
x=house_prise_data.drop(['Price'], axis=1)
y=house_prise_data['Price']
```

```python
print(x)
print(y)
```

Splitting the data in training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model training using XGBoost Regressor

```python
model=XGBRegressor()
```

```python
#training the model with train data
model.fit(x_train, y_train)
```

```python
#accuracy on the train data
taraining_data_prediction = model.predict(x_train)
```

```python
print(taraining_data_prediction)
```

```python
# R square error test
score_1 = metrics.r2_score(y_train, taraining_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(y_train, taraining_data_prediction)

print("R square error : ", score_1)
print('Mean absolute error : ', score_2)
```

Prediction on test data

```python
#accuracy on the test data
testing_data_prediction = model.predict(x_test)
```

```python
# R square error test on test data
score_1 = metrics.r2_score(y_test, testing_data_prediction)

# Mean absolute error on test data
score_2 = metrics.mean_absolute_error(y_test, testing_data_prediction)

print("R square error : ", score_1)
print('Mean absolute error : ', score_2)
```

Visulizing the actual price and predicted price

```python
plt.figure(figsize=(6,6))

plt.scatter(y_test.values, testing_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices (Test Data)")

# reference line (perfect prediction)
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--'
)

plt.show()
```
